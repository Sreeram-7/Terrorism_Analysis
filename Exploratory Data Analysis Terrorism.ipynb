```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset from the local file system
file_path = "C:/Users/hp/OneDrive/Desktop/VCU/Sparks Foundation/Global Terrorism - START data - Copy/globalterrorismdb_0718dist.csv" 
data = pd.read_csv(file_path, encoding='ISO-8859-1')
print("Data loaded successfully!")
print(data.head())
```

```python
# Basic information about the dataset
print(data.info())

# Summary statistics
print(data.describe())

# Checking for missing values
print(data.isnull().sum())
```

```python
# Filling missing values if necessary
data.fillna({'city': 'Unknown', 'summary': 'No summary'}, inplace=True)
```

```python
plt.figure(figsize=(10, 6))
data['iyear'].value_counts().sort_index().plot(kind='bar')
plt.title('Number of Terrorist Attacks per Year')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
data['country_txt'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Countries with Most Terrorist Attacks')
plt.xlabel('Country')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
data['attacktype1_txt'].value_counts().plot(kind='bar')
plt.title('Most Common Types of Attacks')
plt.xlabel('Attack Type')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
data.groupby('iyear')['nkill'].sum().plot(kind='bar')
plt.title('Number of Casualties per Year')
plt.xlabel('Year')
plt.ylabel('Number of Casualties')
plt.show()
```
